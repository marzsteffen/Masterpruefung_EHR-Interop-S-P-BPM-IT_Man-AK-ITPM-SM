```python
# save the base url of the FHIR server
#base = 'http://diam.fh-joanneum.at:8021/hapi-fhir-jpaserver/fhir'
base = 'http://hapi.fhir.org/baseR4'

# setup imports
import os
from requests import get
from requests import post
from requests import put
from requests import delete
from requests import head
from IPython.display import IFrame

# a function to print the top x rows and add a newline
def peek(string, line_count=25):
    print(os.linesep.join(string.split(os.linesep)[:line_count]) + '\n')
```

```python
# HL7 FHIR defines a set of "resources" for exchanging information.
IFrame('https://www.hl7.org/fhir/resourcelist.html#tabs', width=1200, height=330)
```

**#1: What does the "N" on a resource mean?**

*Normativ – Die Resource ist ein stabiler, offiziell verabschiedeter Standard.*

**#2: What does the number on a resource mean?**

*Die Zahl gibt den Reifegrad (Maturity Level) der Resource an.*

**#3: Which are the resources meant for use in clinical research?**


Typische Ressourcen: ResearchStudy, ResearchSubject, ResearchDefinition, ResearchElementDefinition, Evidence, EvidenceVariable, EvidenceReport.*

**#4: How does one define a specific (e.g.) lab test (not taken yet, no patient yet)**

*Mit einer ServiceRequest-Resource, in der der Test als code angegeben wird, aber ohne Patientenzuordnung.*

**#5: How does FHIR code "weight" in an "Observation Resource"?**

*Das Gewicht wird als Observation-Resource mit dem Code für Gewicht (z.B. LOINC: 29463-7) und dem Wert im Feld valueQuantity dokumentiert.*

**#6: Are there other coding systems used?**


*Ja, z.B. LOINC, SNOMED CT, ICD-10, UCUM.*

**#7: Which operations are available?**


*Standard-REST-Operationen: read, vread, update, delete, history, create, search, batch, transaction, operation.*

**#8: How is security handled?**

*Über HTTPS, OAuth2, SMART-on-FHIR, API-Keys, Benutzer-Authentifizierung und Autorisierung.*

**#9: What return formats are supported?**

*JSON (application/fhir+json), XML (application/fhir+xml).*

**#10: What is returned when a resource is not found?**

*HTTP-Statuscode 404 und eine FHIR-OperationOutcome-Resource mit Fehlerbeschreibung.*

**#11: Look for all patients with the name _Andrew_**

```python
response = get(base + '/Patient?given=Andrew')
print('Response code:', response.status_code)
peek('Response body:\n' + response.text)
print('Number of entries:', len(response.json().get('entry', [])))
```

**#12: Look for the patient with ID 697701**

```python
response = get(base + '/Patient/697701')
print('Response code:', response.status_code)
peek('Response body:\n' + response.text)
```

**#13 Look for all the "male"/"female" patients**


```python
print('male:', get(base + '/Patient?gender=male&_summary=count').json().get('total'))
print('female:', get(base + '/Patient?gender=female&_summary=count').json().get('total'))
```

```python
response = get(base + '/Patient?birthdate=lt1950')
peek('Response body:\n' + response.text)
```

**#18 Look for the allergies of patient 1707602**


```python
base = 'http://hapi.fhir.org/baseDstu3'
response = get(base + '/AllergyIntolerance?patient=1707602')
peek('Response body:\n' + response.text)
```

**#19 What substance is the patient allergic to?**

**#20 What category is that?**


```python
*Im Feld category der AllergyIntolerance-Resource nachsehen (z.B. food, medication, environment).*
```

**#21 Look for all allergies of patient 1707602 of which the status is active** 


```python
response = get(base + '/AllergyIntolerance?patient=1707602&clinical-status=active')
peek('Response body:\n' + response.text)
```
